# Which three years of the 1930s saw the most films released?

```python
# Import the necessary libraries
import pandas as pd

# Load the movie titles data from the CSV file
titles = pd.read_csv('titles.csv')

# Filter titles for the 1930s
titles_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]

# Count the number of films released in each year
films_per_year = titles_1930s['year'].value_counts()

# Get the three years with the most films released
top_three_years = films_per_year.head(3)

# Display the top three years
top_three_years
```

# Plot the number of films that have been released each decade over the history of cinema

```python
# Import the necessary libraries for plotting
import matplotlib.pyplot as plt

# Group titles by decade and count the number of films in each decade
titles['decade'] = (titles['year'] // 10) * 10
films_per_decade = titles['decade'].value_counts().sort_index()

# Plot the number of films per decade
films_per_decade.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.show()
```

# How many leading (n=1) roles were available to actors, and how many to actressesin the 1950s?


```python
# Filter cast data for leading roles (n=1) in the 1950s
leading_roles_1950s = cast[(cast['n'] == 1) & (cast['year'] >= 1950) & (cast['year'] <= 1959)]

# Count the number of leading roles for actors and actresses
leading_roles_actors = leading_roles_1950s[leading_roles_1950s['type'] == 'actor']
leading_roles_actresses = leading_roles_1950s[leading_roles_1950s['type'] == 'actress']

# Display the counts
num_leading_roles_actors = len(leading_roles_actors)
num_leading_roles_actresses = len(leading_roles_actresses)

(num_leading_roles_actors, num_leading_roles_actresses)
```
