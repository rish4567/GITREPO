# 1. Write a python program to create a base class "Shape" with methods to calculate area and perimeter Then, create derived classes "Circle" and "Rectangle" that inherit from the base class and calculate their respective areas and perimeters. Demonstrate their usage in a program.

You are developing an online quiz application where users can take quizzes on various topics and receive scores.

```python
import math

class Shape:
    def area(self):
        pass
    
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius**2
    
    def perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

# Demonstrate their usage
circle = Circle(5)
print("Circle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())

rectangle = Rectangle(4, 6)
print("Rectangle Area:", rectangle.area())
print("Rectangle Perimeter:", rectangle.perimeter())
```

```python
class Question:
    def __init__(self, text, correct_answer):
        self.text = text
        self.correct_answer = correct_answer

class Quiz:
    def __init__(self, name, questions):
        self.name = name
        self.questions = questions

    def calculate_score(self, user_answers):
        correct_count = 0
        for i in range(len(self.questions)):
            if user_answers[i] == self.questions[i].correct_answer:
                correct_count += 1
        return (correct_count / len(self.questions)) * 100

# Example usage
question1 = Question("What is 2 + 2?", "4")
question2 = Question("What is the capital of France?", "Paris")

quiz = Quiz("Math and Geography", [question1, question2])

user_answers = ["4", "Paris"]
user_score = quiz.calculate_score(user_answers)
print(f"User's Score: {user_score}%")
```

# 2. Write a python script to create a class "Person" with private attributes for age and name. Implement a method to calculate a person's eligibility for voting based on their age. Ensure that age cannot be accessed directly but only through a getter method.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute for name
        self.__age = age    # Private attribute for age
    
    # Getter method for age
    def get_age(self):
        return self.__age

    def is_eligible_to_vote(self):
        if self.__age >= 18:
            return True
        else:
            return False

    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Age: {self.get_age()}")
        if self.is_eligible_to_vote():
            print("Eligible to vote")
        else:
            print("Not eligible to vote")

# Create a Person object
person1 = Person("Alice", 25)

# Access attributes and methods
person1.display_info()
```

The system should be able to handle different types of accounts, such as Savings Accounts and Checking AccountsBoth account types should have common attributes like an account number, account holder's name, and balance. However, Savings Accounts should have an additional attribute for interest rate, while Checking Accounts should have an attribute for overdraft limit.

1.Create a Python class called Bank Account with the following attributes and methods:

a. Attributes: account_number, account_holder_name, balance

b. Methods: init_()(constructor), deposit()and withdraw()

2.Create two subclasses, Savings Account and CheckingAccountthat inherit from the BankAccount class.

ii)Method: calculate_interest()which calculates and adds interest to the account based on the interest rate.

b. Checking Account:

i)Additional attribute: overdraft_limit

ii)Method: withdraw()which allows withdrawing money up to the overdraft limit (if available) without additional fees.

4.Write a program that creates instances of both Savings Account and Checking Account and demonstrates the use of their methods.

5.Implement proper encapsulation by making the attributes private where necessary and providing getter and setter methods as needed. 

6.Handle any potential errors or exceptions that may occur during operations like withdrawals, deposits, or interest calculations.

7.Provide comments in your code to explain the purpose of each class, attribute and method.

Note: Your code should create instances of the classes, simulate transactions, and showcase the differences between Savings Accounts and Checking Accounts


```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        """
        Constructor for BankAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance (default is 0.0).
        """
        self.__account_number = account_number  # Private attribute
        self.__account_holder_name = account_holder_name  # Private attribute
        self.__balance = balance  # Private attribute

    def get_account_number(self):
        return self.__account_number

    def get_account_holder_name(self):
        return self.__account_holder_name

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        """
        Deposit money into the account.

        Args:
            amount (float): The amount to deposit.

        Returns:
            float: The new balance after deposit.
        """
        if amount > 0:
            self.__balance += amount
            return self.__balance
        else:
            raise ValueError("Deposit amount must be greater than 0")

    def withdraw(self, amount):
        """
        Withdraw money from the account.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            float: The new balance after withdrawal.
        """
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            return self.__balance
        else:
            raise ValueError("Invalid withdrawal amount")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, interest_rate=0.0):
        """
        Constructor for SavingsAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance (default is 0.0).
            interest_rate (float, optional): The interest rate (default is 0.0).
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__interest_rate = interest_rate  # Private attribute

    def get_interest_rate(self):
        return self.__interest_rate

    def calculate_interest(self):
        """
        Calculate and add interest to the account based on the interest rate.

        Returns:
            float: The new balance after adding interest.
        """
        interest = self.__balance * (self.__interest_rate / 100)
        self.__balance += interest
        return self.__balance

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, overdraft_limit=0.0):
        """
        Constructor for CheckingAccount class.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): The initial balance (default is 0.0).
            overdraft_limit (float, optional): The overdraft limit (default is 0.0).
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__overdraft_limit = overdraft_limit  # Private attribute

    def get_overdraft_limit(self):
        return self.__overdraft_limit

    def withdraw(self, amount):
        """
        Withdraw money from the account up to the overdraft limit (if available).

        Args:
            amount (float): The amount to withdraw.

        Returns:
            float: The new balance after withdrawal.
        """
        if amount > 0 and amount <= (self.__balance + self.__overdraft_limit):
            self.__balance -= amount
            return self.__balance
        else:
            raise ValueError("Invalid withdrawal amount")

# Example usage:
try:
    # Create a Savings Account
    savings_account = SavingsAccount("SA123", "Alice", balance=1000.0, interest_rate=5.0)
    print(f"Account Holder: {savings_account.get_account_holder_name()}")
    print(f"Initial Balance: ${savings_account.get_balance():.2f}")
    #savings_account.calculate_interest()
    print(f"Balance after interest: ${savings_account.get_balance():.2f}")

    # Create a Checking Account
    checking_account = CheckingAccount("CA456", "Bob", balance=500.0, overdraft_limit=200.0)
    print(f"Account Holder: {checking_account.get_account_holder_name()}")
    print(f"Initial Balance: ${checking_account.get_balance():.2f}")
    #checking_account.withdraw(700.0)
    print(f"Balance after withdrawal: ${checking_account.get_balance():.2f}")

except ValueError as e:
    print(f"Error: {e}")
```

```python
class Employee:
    def __init__(self, name, employee_id, salary):
        self.__name = name  # Private attribute for name
        self.__employee_id = employee_id  # Private attribute for employee ID
        self.__salary = salary  # Private attribute for salary

    # Getter methods
    def get_name(self):
        return self.__name

    def get_employee_id(self):
        return self.__employee_id

    def get_salary(self):
        return self.__salary

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_employee_id(self, employee_id):
        self.__employee_id = employee_id

    def set_salary(self, salary):
        self.__salary = salary

    def calculate_salary(self):
        """
        Calculate the salary.

        This method should be overridden in subclasses.
        """
        pass

class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, annual_salary):
        super().__init__(name, employee_id, annual_salary)

    def calculate_salary(self):
        """
        Calculate the salary for full-time employees (annual salary).

        Returns:
            float: The annual salary.
        """
        return self.get_salary()

class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate, hours_worked):
        super().__init__(name, employee_id, hourly_rate)
        self.__hours_worked = hours_worked  # Private attribute for hours worked

    # Getter and setter method for hours worked
    def get_hours_worked(self):
        return self.__hours_worked

    def set_hours_worked(self, hours_worked):
        self.__hours_worked = hours_worked

    def calculate_salary(self):
        """
        Calculate the salary for part-time employees (hourly rate * hours worked).

        Returns:
            float: The monthly salary.
        """
        return self.get_salary() * self.__hours_worked

# Example usage:
full_time_employee = FullTimeEmployee("John", "FT001", 60000.0)
part_time_employee = PartTimeEmployee("Alice", "PT001", 20.0, 120)

# Calculate and display salaries
print(f"{full_time_employee.get_name()} (Employee ID: {full_time_employee.get_employee_id()})")
print(f"Annual Salary: ${full_time_employee.calculate_salary():.2f}\n")

print(f"{part_time_employee.get_name()} (Employee ID: {part_time_employee.get_employee_id()})")
print(f"Monthly Salary: ${part_time_employee.calculate_salary():.2f}")
```

```python
# Define a base class for LibraryItems (books, DVDs, etc.)
class LibraryItem:
    def __init__(self, title, item_id):
        self.__title = title  # Private attribute for item title
        self.__item_id = item_id  # Private attribute for item ID
        self.__checked_out = False  # Private attribute for item status (checked out or not)
    
    def get_title(self):
        return self.__title
    
    def get_item_id(self):
        return self.__item_id
    
    def is_checked_out(self):
        return self.__checked_out
    
    def check_out(self):
        self.__checked_out = True
    
    def return_item(self):
        self.__checked_out = False

# Create subclasses for different types of LibraryItems
class Book(LibraryItem):
    def __init__(self, title, item_id, author, genre):
        super().__init__(title, item_id)
        self.__author = author  # Private attribute for book author
        self.__genre = genre  # Private attribute for book genre
    
    def get_author(self):
        return self.__author
    
    def get_genre(self):
        return self.__genre

class Patron:
    def __init__(self, patron_id, name):
        self.__patron_id = patron_id  # Private attribute for patron ID
        self.__name = name  # Private attribute for patron name
        self.__checked_out_items = []  # Private attribute to track checked out items
    
    def get_patron_id(self):
        return self.__patron_id
    
    def get_name(self):
        return self.__name
    
    def get_checked_out_items(self):
        return self.__checked_out_items
    
    def check_out_item(self, item):
        if item.is_checked_out():
            print("Item is already checked out.")
        else:
            item.check_out()
            self.__checked_out_items.append(item)
    
    def return_item(self, item):
        if item in self.__checked_out_items:
            item.return_item()
            self.__checked_out_items.remove(item)
        else:
            print("Item is not checked out to this patron.")
    
    def view_checked_out_items(self):
        for item in self.__checked_out_items:
            print(f"Title: {item.get_title()}, Item ID: {item.get_item_id()}")

# Example usage:
book1 = Book("The Great Gatsby", "B001", "F. Scott Fitzgerald", "Fiction")
book2 = Book("To Kill a Mockingbird", "B002", "Harper Lee", "Fiction")

patron1 = Patron("P001", "Alice")
patron2 = Patron("P002", "Bob")

patron1.check_out_item(book1)
patron2.check_out_item(book2)

patron1.view_checked_out_items()
patron2.view_checked_out_items()
```
